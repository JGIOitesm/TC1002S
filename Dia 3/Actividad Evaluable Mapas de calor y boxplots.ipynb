### Carga los datos usando tu lector de csv o con pandas. Es recomendable hacerlo con pandas.

```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('covid19_tweets.csv')
```

### Realiza el análisis de las variables usando diagramas de cajas y bigotes, histogramas y mapas de calor.

```python
plt.boxplot(df[["user_followers","user_friends","user_favourites"]])

plt.show()
```

```python
ax = df[["user_followers","user_friends","user_favourites"]].plot.hist(bins=10, alpha=0.5)
```

```python
ax = sb.heatmap(df[["user_followers","user_friends","user_favourites"]])
```

```python
details = pd.DataFrame(df[["user_followers","user_friends","user_favourites"]])

details = details.apply(lambda x : True if x['user_followers'] <1 and x['user_favourites']<1  else False, axis = 1)
  
num_rows = len(details[details == True].index)
  
print('Cuentas sin seguidores ni usuarios seguidos : ',
      num_rows )
```

```python
df.describe(include = object).transpose()
```

```python
details = pd.DataFrame(df[["user_followers","user_friends","user_favourites","user_verified"]])
  

details = details.apply(lambda x : True
            if x['user_verified']==True   else False, axis = 1)
  
# Count number of True in the series
num_rows = len(details[details == True].index)
  
print('Cuentas verificadas: ',
      num_rows )
```

```python
details = pd.DataFrame(df[["user_followers","user_friends","user_favourites","user_verified"]])
  

details = details.apply(lambda x : True
            if x['user_verified']==False   else False, axis = 1)
  
# Count number of True in the series
num_rows2 = len(details[details == True].index)
  
print('Cuentas no verificadas: ',
      num_rows2 )
```

```python
df1 = pd.DataFrame({'Usuarios':['Verificado', 'No_verificado'], 'Cantidad':[num_rows,  num_rows2 ]})
ax = df1.plot.bar(x='Usuarios', y='Cantidad', rot=0,color={"red"})
```
