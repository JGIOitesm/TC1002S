### Carga los datos usando tu lector de csv o con pandas. Es recomendable hacerlo con pandas.

```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

df = pd.read_csv('covid19_tweets.csv')
```

### Realiza el análisis de las variables usando diagramas de cajas y bigotes, histogramas y mapas de calor.

```python
plt.boxplot(df[["user_followers","user_friends","user_favourites"]])

plt.show()
```

```python
ax = df[["user_followers","user_friends","user_favourites"]].plot.hist(bins=10, alpha=0.5)
```

```python
ax = sb.heatmap(df[["user_followers","user_friends","user_favourites"]])
```
